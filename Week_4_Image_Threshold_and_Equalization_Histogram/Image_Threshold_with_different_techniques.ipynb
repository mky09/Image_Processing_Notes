```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img_rgb = cv2.imread("./pngtree-river-in-an-autumn-forest-with-a-sun-coming-through-the-image_2903957.jpg",1)
my_img_gray = cv2.cvtColor(my_img_rgb, cv2.COLOR_BGR2GRAY)
```

```python
my_thresh_binary = cv2.threshold(my_img_gray,thresh= 128 , maxval=255,
                                 type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
rows= 1 
columns = 3

fig.add_subplot(rows, columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("orginal mg")

fig.add_subplot(rows, columns,2)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("BIARY THRESH IMAGE")


fig.add_subplot(rows, columns,3)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("BIARY INV THRESH IMAGE")
```

```python
my_thresh_trunc = cv2.threshold(my_img_gray,thresh= 128 , maxval=255,
                                 type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(10,7))
rows= 1 
columns = 4

fig.add_subplot(rows, columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("orginal mg")

fig.add_subplot(rows, columns,3)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("BIARY THRESH IMAGE")


fig.add_subplot(rows, columns,4)
plt.imshow(my_thrsh_trunc,cmap="gray")
plt.axis("off")
plt.title("BIARY INV THRESH IMAGE")
```
