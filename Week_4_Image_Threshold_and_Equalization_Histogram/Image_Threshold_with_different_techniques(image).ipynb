```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img_rgb= cv2.imread("./pngtree-river-in-an-autumn-forest-with-a-sun-coming-through-the-image_2903957.jpg",1)
my_img_gray= cv2.cvtColor(my_img_rgb,cv2.COLOR_BGR2GRAY)
```

```python
_, my_tresh_binary= cv2.threshold(my_img_gray, thresh=128,maxval=255,
                                  type=cv2.THRESH_BINARY)
```

```python
fig=plt.figure(figsize=(10,7))
rows = 1
columns = 2
fig.add_subplot(rows, columns,1)
plt.imshow(my_img_gray,cmap="gray")
plt.axis("off")
plt.title("original mage")
fig.add_subplot(rows,columns,2)
plt.imshow(my_tresh_binary,cmap="gray")
plt.axis("off")
plt.title("binary")
```
